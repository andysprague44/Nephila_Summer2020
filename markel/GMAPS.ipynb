## LOGIC:
 ### Most of the google api types seem to be commercial, and if not commercial then it is related to government, religion/nonprofit, restaurant, transportation.
 
 1. If **air code** == 300: it assumed to be unknown, and that unknown is the correct value
 2. If **air code** any of **'residential'**, which is 301 - 306, then they must not have any of the types in google_types. 
         - EXCEPTION: 'spa' and 'lodging' (gmaps) falls under 304(AIR) temporary housing.
         - EXCEPTION: names with 'Hotel' , 'Condo' or 'Resort' in AIR may not be encoded as spa/lodging/realestate agency/travel agency but 'establishment' in gmaps. Accounted for by checking if name contains Hotel or resort
 5. If **google maps** code is restaurant, then air code should be 331- restaurants. 
 6. If **google maps** code is gas_station, then AIR should also be 335-gas station.
 7. If **google maps** code is 'car_wash', then AIR should also be 336- Automotive repair shops and Car Washes
 8. If **google maps** is any of the following, AIR should be either 341 or 342, (Religion and Non-Profit) 
     - church
     - hindu_temple
     - mosque
     - synagogue
   
 9. If **google maps** is any of the following, AIR should be 343-Government General Services
     - city_hall
     - courthouse
     - embassy
     - library
     - local_government_office
     - police
     - post_office (may also be 344)
   
 10. if **google maps** is any of the following, AIR should be 344- Governement Emergency Services
     - fire_station 
     - hospital (may also be listed as commercial health services)
     - police (may also be 343)
    
 11. If **google maps** is any type of 'school':
     - school --> air 345 or 346
     - university --> air 345
     - primary_school --> air 346
     - secondary school --> air 346
 12. If **google maps** is light_rail_station, subway_station, train_station, it should be AIR code 352 - Railroad
 13. If **google maps** is airport, then air code should be 353 - Air.
 12. If **air code** is any of **'commercial'** , 312 - 319, OR if it is encoded as 'general commercial' which is 311, then it should have a google type that is **not** specified in the manual mappings in _steps 3-11_. 'General Commercial' (AIR) will also be synonymous to 'establishment'(gmaps)
 
 12. Google maps does not cover types between 321-330 (Industrial), 361-365 (Utilities), 371-381 (Construction/Erection Risks), and 400-482 (Industrial Facility), so we will sample a subset of records with these occupancy codes and *manually check them via google maps.* and estimate the error for these.
 
 
 Google maps ambiguous type tags interpretations:
 - establishment -> general building, which can be tagged for hotels, condos, etc, but also non condos, hotels, etc.
 - premise -> should be 300 or 'unknown'.
 - street_address -> no commercial tagging, just lot or residential
